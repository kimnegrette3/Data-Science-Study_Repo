```python
# Import modules
import pandas as pd                 #for dataframe and series analysis
import matplotlib.pyplot as plt     #for data visualization
import seaborn as sns               #for data visualization
import datetime as dt               #for datetime data type
```

## Read dataset

```python
# Import csv file 
#You can indicate to pandas how to identify the null values in your data
trips = pd.read_csv('trips_combined.csv', na_values='null') 
#With head you quickly check how your columns and first rows look like
trips.head()                                               
```

```python
#Get to know more about the dataset. With info() you get the total number of observations, columns, 
#the non-null count and the data type of each column. 
#This way you can spot incorrect datatypes and correct them
trips.info(show_counts=True)
```

```python
# Modify data types so all dates are in the correct datetime format
trips['started_at'] = pd.to_datetime(trips['started_at'], format='%Y-%m-%d %H:%M:%S')
trips['ended_at'] = pd.to_datetime(trips['ended_at'], format='%Y-%m-%d %H:%M:%S')
```

```python
#Verify that the changes were made
trips.info()
```

```python
#Get some preliminary summary descriptions using the describe() method. 
#You can pass the argument include=['object] to get the summary for non-numeric columns
summary_objects = pd.DataFrame(trips.describe(include=['object']))
summary_objects
```

```python
#But by default this method gives you the summary for numeric columns
summary_numbers = pd.DataFrame(trips.describe())
summary_numbers
```

## Charts

### Trip count by month

```python
# Trip count by user type and month of the year using a grouped bar graph and the seaborn module
# You indicate the type of graph with 'kind', and the variable you want to group by using 'hue'. 
# The params 'height', 'aspect', 'alpha' and 'palette' are for plot dimensions, transparency and color
p = sns.catplot(data=trips, kind='count', x='started_month',
                 hue='user_type', height=6, aspect=1.8, alpha=0.9, palette='Accent')
p.set_xticklabels(rotation=30, ha='right') #Rotate the labels in x axis and align them to the right
p.set_axis_labels("Month of the Year", "Total rides", fontsize=14)
p._legend.set_title('User type')
p.fig.subplots_adjust(top=0.9), # Set the distance between tittle and figure
p.fig.suptitle('Ride Count by User Type and Month', 
               fontsize=20, x=0.25, y=1.05) # Set tittle, its size and alignment
```

```python
#Make a pivot table to see the aggregation. It's the same concept as a pivot table in excel! 
#Give the values you want to group ('value') and the variables used to group them ('index' and 'columns').
#Also indicate the aggregation function you need with 'aggfunc'
month_order = ['January', 'February', 'March','April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
pivot_count_bymonth = pd.pivot_table(trips, values='ride_id', index='user_type', 
                                     columns='started_month', aggfunc='count')
pivot_count_bymonth.reindex(columns=month_order)
```

### Trip count by day of the week

```python
# Add a new column to the dataframe with the date only, to aggregate day of the week later
# Here I'm using the 'dt' method form datetime module 
trips['started_date'] = trips['started_at'].dt.date
trips.head()
```

```python
# To plot average number of trips for each day of the week, we need to make a groupby to calculate the count of each one. 
days = pd.DataFrame(trips.groupby(['user_type','started_day','started_date']).size().reset_index(name='count'))
days
```

```python
# Plot the average ride count using a point graph
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
x = sns.catplot(data=days, kind='point', x='started_day', y='count', hue='user_type',
                 errorbar=None, order=order, height=6, aspect=1.8, palette='Accent')
x.set_axis_labels("Day of the Week", "Average ride count", fontsize=14)
x._legend.set_title('User type')
x.fig.subplots_adjust(top=0.9),
x.fig.suptitle('Ride Count by User Type and Day of the Week', fontsize=20, x=0.25, y=1.05)
```

```python
#Make a pivot table from the created groupby to see the aggregation for each user type
pivot_count_byday = pd.pivot_table(days, values='count', index='user_type', 
                                   columns='started_day', aggfunc='mean')
pivot_count_byday.reindex(columns=order).round(2)
```

### Trip count by hour

```python
# Group by hours of the day
# Extract hour from column started_at and create a new column in the dataframe
trips['started_hour'] = trips['started_at'].dt.hour
trips.head()
```

```python
#Group by hours to get the count of each one
hours = pd.DataFrame(trips.groupby(['user_type','started_day','started_hour']).size().reset_index(name='count'))
hours
```

```python
# Plot the average ride count by hour using a point graph
h = sns.catplot(data=hours, x='started_hour', y='count', kind='point', hue='user_type', 
                errorbar=None, height=6, aspect=1.8, palette='Accent')
h.set_axis_labels("Hour", "Average ride count", fontsize=14)
h._legend.set_title('User type')
h.fig.subplots_adjust(top=0.9),
h.fig.suptitle('Ride Count by User Type and Hour of the day', fontsize=20, x=0.25, y=1.05)
```

```python
# Adding a categorical column to the previously created table, indicating weekday or weekend
hours['day_type'] = hours['started_day'].apply(lambda x: 'Weekday' 
                                               if x in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'] 
                                               else 'Weekend')
```

```python
hours.head()
```

```python
#Plot the average ride count by hour during the weekdays and weekends
#To do this we add the 'col' parameter to the sns.catplot, 
#that way we tell seaborn to separate the data using the variable 'day_type'
g = sns.catplot(data=hours, x='started_hour', y='count', kind='point', hue='user_type', 
                errorbar=None, height=6, aspect=1.1, col='day_type', palette='Accent')
(g.set_axis_labels("Hour", "Average ride count", fontsize=14)
  .set_titles("{col_name}")
  .despine(left=True))  
g._legend.set_title('User type')
g.fig.subplots_adjust(top=0.9),
g.fig.suptitle('Ride Count by User Type and Hour', fontsize=20, x=0.25, y=1.05)
```

```python
#Make a pivot table to see the aggregation 
pivot_count_byhour = pd.pivot_table(hours, values='count', index='started_hour', 
                                    columns=['day_type', 'user_type'], aggfunc='mean')
pivot_count_byhour.round(2)
```

### Trip duration by month

```python
# Trip duration by user type and month
u = sns.catplot(data=trips, kind='bar', x='started_month', y='trip_duration', hue='user_type', 
                errorbar=None, height=6, aspect=1.8, alpha=0.9, palette='Accent')
u.set_xticklabels(rotation=30, ha='center', fontsize=12)
u.set_axis_labels("Month of the Year", "Average Trip Duration (minutes)", fontsize=14)
u._legend.set_title('User type')
u.fig.subplots_adjust(top=0.9),
u.fig.suptitle('Average Trip Duration by User Type and Month', fontsize=20, x=0.25, y=1.05)
```

```python
#Make a pivot table to see aggregation
pivot_dur_bymonth = pd.pivot_table(trips, values='trip_duration', index='user_type',
                                    columns='started_month', aggfunc='mean')
pivot_dur_bymonth.reindex(columns=month_order).round(2)
```

### Trip duration by day of the week

```python
# Trip duration by user type and day of the week
v = sns.catplot(data=trips, kind='point', x='started_day', y='trip_duration', 
                hue='user_type', order=order,
                errorbar=None, height=6, aspect=1.8, palette='Accent')
v.set_xticklabels(rotation=0, ha='center', fontsize=12)
v.set_axis_labels("Day of the Week", "Average Trip Duration (minutes)", fontsize=14)
v._legend.set_title('User type')
v.fig.subplots_adjust(top=0.9),
v.fig.suptitle('Average Trip Duration by User Type and Day of the Week', 
               fontsize=20, x=0.30, y=1.05)
```

## Insights
Here are some quick insights we can extract from the tables and graphs made using pandas.

1. During the summer months, there is an increase in the number of trips.
2. On Fridays, Saturdays, and Sundays, there is an increase in the number of trips by casual users.
3. On the other hand, there is a decrease in the number of trips by members on Fridays, Saturdays, and Sundays.
4. During the week, there is an increase in the number of trips by member users during peak hours, specifically between 7-8am and 5-6pm.
5. The average duration of trips for member users remains relatively stable throughout the year. However, for casual users, it reaches its peak in the spring months (from April to June) and then begins to decrease.
6. The average duration of trips varies little for member users throughout the week. However, in the case of casual users, an increase of up to 5 minutes is observed on weekends.
7. During weekdays, from Monday to Friday, the average duration of trips for member users does not show significant variations through the day. For casual users, the average duration reaches its minimum at around 7am and its maximum between 10am and 12pm.
8. On Saturdays and Sundays, member users show a stable behavior throughout the day. In contrast, casual users show an increase in the duration of trips starting at 7 am, reaching a peak at 1 pm, and then starting to decrease.
